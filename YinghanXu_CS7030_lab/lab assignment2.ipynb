```python
%matplotlib inline 

import numpy as np
import pylab
import matplotlib.pyplot as plt

from scipy.fftpack import rfft, rfftfreq, irfft

def setup_graph(title='', x_label='', y_label='', fig_size=None):
    fig = plt.figure()
    if fig_size != None:
        fig.set_size_inches(fig_size[0], fig_size[1])
    ax = fig.add_subplot(111)
    ax.set_title(title)
    ax.set_xlabel(x_label)
    ax.set_ylabel(y_label)

def rect(x):
    rectOut = np.zeros(len(x))
    for i in range(len(x)):
        if np.abs(x[i]) > 1/2:
            rectOut[i] = 0
        elif np.abs(x[i] == 1/2):
            rectOut[i] = 1/2
        elif np.abs(x[i]) < 1/2:
            rectOut[i] = 1
    return rectOut

def triangle(x):
    triOut = np.zeros(len(x))
    for i in range(len(x)):
        if np.abs(x[i]) < 1:
            triOut[i] = 1 - np.abs(x[i])
        else:
            triOut[i] = 0
    return triOut

def impulse(x):
    impOut = np.zeros(len(x))
    for i in range(len(x)):
        if x[i] == 0:
            impOut[i] = 1
        else:
            impOut[i] = 0
    return impOut

time_to_plot = 20
sample_rate = 1000
t = np.linspace(-20, time_to_plot, sample_rate * time_to_plot)

#The rectangle function

#%%
signal_rec = rect(t)

setup_graph(x_label='time (in seconds)', y_label='amplitude', title='time domain')

plt.plot(t, signal_rec)
plt.show()

#%%
fft_out_rec = rfft(signal_rec)
fft_magnitude_rec_only = np.abs(fft_out_rec)
frequencies_fft_rec = rfftfreq(len(signal_rec), 1/sample_rate)

setup_graph(x_label='frequency (in Hz)', y_label='amplitude', title='frequency domain')

#plt.plot(frequencies_fft_rec, fft_magnitude_rec_only, 'r')
print(fft_out_rec)
plt.plot(frequencies_fft_rec, fft_out_rec, 'r')
plt.show()
plt.plot(frequencies_fft_rec, fft_magnitude_rec_only, 'g')
plt.show()

#%%
power_rec = (fft_magnitude_rec_only)**2

setup_graph(x_label='frequency (in Hz)', y_label='power', title='power spectra')

plt.plot(frequencies_fft_rec, power_rec, 'g')
plt.show()

#The sinc function

#%%

signal_sinc = np.sinc(t)

setup_graph(x_label='time (in seconds)', y_label='amplitude', title='time domain')

plt.plot(t, signal_sinc)
plt.show()

#%%
fft_output_sinc = rfft(signal_sinc)
fft_magnitude_sinc_only = np.abs(fft_output_sinc)
frequencies_fft_sinc = rfftfreq(len(signal_sinc), 1/sample_rate)

setup_graph(x_label='frequency (in Hz)', y_label='amplitude', title='frequency domain')

plt.plot(frequencies_fft_sinc, fft_magnitude_sinc_only, 'r')
plt.show()

#%%
power_sinc = (fft_magnitude_sinc_only)**2

setup_graph(x_label='frequency (in Hz)', y_label='power', title='power spectra')

plt.plot(frequencies_fft_sinc, power_sinc, 'g')
plt.show()


#The signum function
#%%
signal_signum = np.sign(t)

setup_graph(x_label='time (in seconds)', y_label='amplitude', title='time domain')

plt.plot(t, signal_signum)
plt.show()

#%%
fft_output_signum = rfft(signal_signum)
fft_magnitude_signum_only = np.abs(fft_output_signum)
frequencies_fft_signum = rfftfreq(len(signal_signum), 1/sample_rate)
setup_graph(x_label='frequency (in Hz)', y_label='amplitude', title='frequency domain')

plt.plot(frequencies_fft_signum, fft_magnitude_signum_only, 'r')
plt.show()

#%%
power_signum = np.square(np.abs(fft_output_signum))

setup_graph(x_label='frequency (in Hz)', y_label='power', title='power spectra')

plt.plot(frequencies_fft_signum, power_signum, 'g')
plt.show()

#The triangle function
#%%
signal_tri = triangle(t)

setup_graph(x_label='time (in seconds)', y_label='amplitude', title='time domain')

plt.plot(t, signal_tri)
plt.show()

#%%
fft_output_tri= rfft(signal_tri)
fft_magnitude_tri_only = np.abs(fft_output_tri)
frequencies_fft_tri = rfftfreq(len(signal_tri), 1/sample_rate)

setup_graph(x_label='frequency (in Hz)', y_label='amplitude', title='frequency domain')

plt.plot(frequencies_fft_tri, fft_magnitude_tri_only, 'r')
plt.show()

#%%
power_tri = (fft_output_tri)**2

setup_graph(x_label='frequency (in Hz)', y_label='power', title='power spectra')

plt.plot(frequencies_fft_tri, power_tri, 'g')
plt.show()

#The impulse delta function
#%%
signal_imp = impulse(t)

setup_graph(x_label='time (in seconds)', y_label='amplitude', title='time domain')

plt.plot(t, signal_imp)
plt.show()

#The comb function
```
